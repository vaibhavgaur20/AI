```python
import pandas

from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
```

```python
dataset = pandas.read_excel("Input.xlsx",engine='openpyxl')
```

```python
url = dataset
```

```python
# import sys
# !{sys.executable} -m pip install webdriver_manager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
driver.get(url)

a = driver.find_elements(By.XPATH,"//h1[contains(@class,'entry-title')]")
print(a[0].text)
print('_'*50)
b = driver.find_elements(By.XPATH,"//p")
for x in b:
    print(x.text)
```
